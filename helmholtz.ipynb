```
\[ScriptCapitalL] = {\[Phi][0, 0]''[r] + 
     2/r \[Phi][0, 0]'[r] - (0 (0 + 1))/
      r^2 \[Phi][0, 0][r] == -\[Lambda] \[Phi][0, 0][r]};
```

```
\[ScriptCapitalB] = {\[Phi][0, 0][1] == 0};
\[ScriptCapitalV] = {\[Phi][0, 0]};
```

```
A = FennecFun[
  Join[\[ScriptCapitalL], \[ScriptCapitalB]], \[ScriptCapitalV], {r, 
   0, 1}, 20, eigenvalue -> \[Lambda]]
```

```
Block[
 {nr = Fennec`nr, n = Fennec`nColTotal, 
  nbc = Length@Fennec`rbc[[1]], \[Delta]},
 \[Delta] = n - nbc;
 p = SparseArray@
   IdentityMatrix[n][[
    Riffle[Range[\[Delta]], Range[\[Delta] + 1, n], {1, n, nr/2}]]]
 ]
```

```
Dot[p, #] & /@ A;
GraphicsRow[MatrixPlot /@ %]
```
