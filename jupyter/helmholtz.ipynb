```
l=2;
equations = {f[l, 0]''[r] + 2/r*f[l, 0]'[r] - (l*(l + 1))/r^2*f[l, 0][r] == -lamb*f[l, 0][r]};
```

```
boundaryconditions = {f[l, 0][1] == 0};
variables = {f[l, 0]};
```

```
A = SproutsFun[Join[equations,boundaryconditions], variables, {r,0,1}, 40, eigenvalue -> lamb]
```

```
GraphicsRow[{MatrixPlot[A[[1]],ImageSize->Small,PlotLabel->"A0"],MatrixPlot[A[[2]],ImageSize->Small,PlotLabel->"A1"]}]
```

```
Sort@Eigenvalues[{A[[1]], -A[[2]]}] // Sqrt
```
